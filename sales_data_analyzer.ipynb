```python
# 1. Importieren der notwendigen Bibliotheken (Библиотеки для работы с данными)
import pandas as pd
import matplotlib.pyplot as plt

# 2. Erstellen von fiktiven Verkaufsdaten (Создаем сырые данные - обычно они приходят из Excel)
data = {
    'Datum': ['2026-03-01', '2026-03-02', '2026-03-03', '2026-03-04', '2026-03-05', '2026-03-06'],
    'Produkt_Kategorie': ['Software-Lizenz', 'Hardware', 'IT-Support', 'Software-Lizenz', 'Hardware', 'Hardware'],
    'Umsatz_EUR': [1500, 850, 300, 1200, 950, 400]
}

# 3. Umwandeln in einen DataFrame (Таблица Pandas) und Speichern als CSV
df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
print("Schritt 1: CSV-Datei erfolgreich generiert!\n")

# 4. Daten einlesen und analysieren (Чтение и группировка данных)
df_read = pd.read_csv('sales_data.csv')
print("Schritt 2: Rohdaten (Sichtprüfung):")
print(df_read.head()) # Показывает первые 5 строк

# Gruppieren nach Produktkategorie und Summe des Umsatzes berechnen
umsatz_pro_kategorie = df_read.groupby('Produkt_Kategorie')['Umsatz_EUR'].sum()
print("\nSchritt 3: Gesamtumsatz pro Kategorie:")
print(umsatz_pro_kategorie)

# 5. Daten visualisieren (Построение графика)
plt.figure(figsize=(8, 5))
umsatz_pro_kategorie.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Gesamtumsatz nach IT-Produktkategorie (März 2026)')
plt.ylabel('Umsatz in EUR')
plt.xlabel('Kategorie')
plt.xticks(rotation=0) # Делает подписи горизонтальными
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Grafik anzeigen
plt.tight_layout()
plt.show()
```
